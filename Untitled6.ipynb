```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

def gabor_features(img, kernels):
    """Extract Gabor features from an image using multiple kernels."""
    features = []
    for kernel in kernels:
        filtered = cv2.filter2D(img, cv2.CV_8UC3, kernel)
        mean = filtered.mean()
        variance = filtered.var()
        features.extend([mean, variance])
    return features

# Step 1: Upload a single image
print("Upload a single image (e.g., texture, face, fingerprint):")
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Step 2: Read the image in grayscale
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if img is None:
    print(f"Error: Could not load image from {image_path}")
else:
    # Step 3: Create Gabor kernels at different orientations and frequencies
    kernel_size = 21
    sigmas = [4.0, 6.0]
    lambdas = [10.0, 15.0]
    gammas = [0.5]
    psis = [0]
    orientations = [0, np.pi/4, np.pi/2, 3*np.pi/4]
    kernels = []
    for sigma in sigmas:
        for lambd in lambdas:
            for gamma in gammas:
                for psi in psis:
                    for theta in orientations:
                        kernel = cv2.getGaborKernel((kernel_size, kernel_size), sigma, theta, lambd, gamma,
                                                psi, ktype=cv2.CV_32F)
                        kernels.append(kernel)

    # Step 4: Extract Gabor features from the uploaded image
    extracted_features = gabor_features(img, kernels)
    print(f"\nExtracted Gabor Features for '{image_path}':")
    print(extracted_features)

    # Optional: Display the original image and some filtered versions for visual inspection
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    if kernels:
        # Apply one of the Gabor filters and display it
        example_filtered_img = cv2.filter2D(img, cv2.CV_8UC3, kernels[0])
        plt.subplot(1, 2, 2)
        plt.imshow(example_filtered_img, cmap='gray')
        plt.title('Example Gabor Filtered Image')
        plt.axis('off')
    plt.show()
```
